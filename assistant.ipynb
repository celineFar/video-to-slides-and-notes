```python
from yt_dlp import YoutubeDL
import logging
import sys

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s %(levelname)s %(message)s",
    stream=sys.stdout,
)

def download_video(url: str, output_dir: str = "./downloads"):
    ydl_opts = {
        # Output template
        "outtmpl": f"{output_dir}/%(title)s.%(ext)s",

        # Reliability
        "retries": 10,
        "fragment_retries": 10,
        "concurrent_fragment_downloads": 1,

        # Avoid being too aggressive
        "sleep_interval": 1,
        "max_sleep_interval": 5,

        # Pretend to be a real browser (NO cookies)
        "user_agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/120.0 Safari/537.36"
        ),

        # Logging
        "quiet": False,
        "no_warnings": False,

        # Format selection (safe default)
        "format": "bv*+ba/best",

        # Network hygiene
        "socket_timeout": 30,

        # Metadata
        "addmetadata": True,
        "writethumbnail": False,
    }

    logging.info("Starting download: %s", url)

    try:
        with YoutubeDL(ydl_opts) as ydl:
            ydl.download([url])
        logging.info("Download completed successfully")
    except Exception as e:
        logging.exception("Download failed")
        raise


test_url = "https://www.youtube.com/watch?v=ecK3EnyGD8o"
download_video(test_url)
```

```python
import os
import time
import re
import yt_dlp
import logging

logger = logging.getLogger(__name__)


def sanitize_filename(filename: str) -> str:
    return re.sub(r"[^\w\-_.]", "_", filename)


def download_if_needed(
    source: str,
    max_resolution: int | None = None,
    video_dir: str = "videos"
    verbose: bool = False,
) -> str | None:
    """
    Download a video if the source is a URL and return a sanitized title.
    """
    if not source.startswith(("http://", "https://", "www.")):
        return source

    video_dir = "videos"
    os.makedirs(video_dir, exist_ok=True)

    title = download_video(
        url=source,
        output_dir=video_dir,
        max_resolution=max_resolution,
        verbose=verbose,
    )

    return os.path.join(video_dir, sanitize_filename(title))


def download_video(
    url: str,
    output_dir: str,
    max_resolution: int | None = None,
    verbose: bool = False,
) -> str:
    """
    Download a video using yt-dlp and return the video title.
    """

    ydl_opts = {
        # 👇 Folder + filename both use the title
        "outtmpl": os.path.join(
            output_dir,
            "%(title)s",
            "%(title)s.%(ext)s",
        ),

        "format": "bestvideo+bestaudio/best",
        "merge_output_format": "mp4",
        "verbose": verbose,

        # IMPORTANT: avoid invalid filename characters (Windows-safe)
        "restrictfilenames": True,

        # Cookies
        "cookiefile": "cookies.txt",

        # AWS / rate-limit friendly throttling
        "sleep_requests": 2,
        "sleep_interval": 2,
        "max_sleep_interval": 6,

        # Browser headers
        "http_headers": {
            "User-Agent": (
                "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                "AppleWebKit/537.36 (KHTML, like Gecko) "
                "Chrome/121.0.0.0 Safari/537.36"
            ),
            "Accept-Language": "en-US,en;q=0.9",
        },
    }

    if max_resolution:
        ydl_opts["format"] = (
            f"bestvideo[height<={max_resolution}]+bestaudio/best"
        )

    max_retries = 3
    last_exception = None

    for attempt in range(1, max_retries + 1):
        try:
            with yt_dlp.YoutubeDL(ydl_opts) as ydl:
                info = ydl.extract_info(url, download=True)
                return info.get("title", "unknown_title")

        except Exception as exc:
            last_exception = exc
            logger.warning(
                "Download attempt %s/%s failed: %s",
                attempt,
                max_retries,
                exc,
            )
            if attempt < max_retries:
                time.sleep(5)

    raise RuntimeError("Failed to download video") from last_exception


# Example usage
url = "https://www.youtube.com/watch?v=ecK3EnyGD8o"
download_if_needed(url)
```

```python
from typing import List, Tuple, Optional
from langchain_community.document_loaders import YoutubeLoader
from langchain_community.document_loaders.youtube import TranscriptFormat


def get_chunked_transcript(
    video_url: str,
    chunk_length: int,
    languages: Optional[List[str]] = None,
    add_video_info: bool = False,
    cookies_path: str | None = "/home/cookies.txt",
) -> List[Tuple[float, float, str]]:

    if languages is None:
        languages = ["zh-Hant", "zh-Hans", "ja", "en"]

    yt_dlp_kwargs = {}
    if cookies_path:
        yt_dlp_kwargs["cookies"] = cookies_path

    loader = YoutubeLoader.from_youtube_url(
        video_url,
        add_video_info=add_video_info,
        transcript_format=TranscriptFormat.CHUNKS,
        chunk_size_seconds=chunk_length,
        language=languages,
        yt_dlp_kwargs=yt_dlp_kwargs,  # ✅ CORRECT FOR YOUR VERSION
    )

    docs = loader.load()

    starts: List[float] = []
    for doc in docs:
        md = doc.metadata
        raw = md.get("start_seconds") or md.get("start") or md.get("start_time")
        try:
            starts.append(float(raw))
        except (TypeError, ValueError):
            starts.append(0.0)

    chunks: List[Tuple[float, float, str]] = []
    for i, doc in enumerate(docs):
        start = starts[i]
        end = starts[i + 1] if i < len(docs) - 1 else start + chunk_length
        chunks.append((start, end, doc.page_content))

    return chunks
```

```python
from typing import List, Tuple, Optional
import subprocess
import tempfile
import os
import pysrt


def get_chunked_transcript(
    video_url: str,
    chunk_length: int,
    languages: Optional[List[str]] = None,
    cookies_path: str | None = "/home/cookies.txt",
) -> List[Tuple[float, float, str]]:
    """
    Download and chunk a YouTube transcript using yt-dlp.

    Returns:
        List of (start_time, end_time, text)
    """

    if languages is None:
        languages = ["en"]

    with tempfile.TemporaryDirectory() as tmpdir:
        output_template = os.path.join(tmpdir, "%(id)s.%(ext)s")

        cmd = [
            "yt-dlp",
            "--skip-download",
            "--write-auto-sub",
            "--write-sub",
            "--sub-format", "srt",
            "--sub-langs", ",".join(languages),
            "-o", output_template,
        ]

        if cookies_path:
            cmd += ["--cookies", cookies_path]

        cmd.append(video_url)

        subprocess.run(
            cmd,
            check=True,
            stdout=subprocess.DEVNULL,
            stderr=subprocess.DEVNULL,
        )

        # find the downloaded srt
        srt_files = [
            f for f in os.listdir(tmpdir)
            if f.endswith(".srt")
        ]

        if not srt_files:
            raise RuntimeError("No subtitles were downloaded")

        srt_path = os.path.join(tmpdir, srt_files[0])
        subs = pysrt.open(srt_path)

        chunks: List[Tuple[float, float, str]] = []

        current_start = None
        current_text = []
        last_end = 0.0

        def to_seconds(t):
            return (
                t.hours * 3600 +
                t.minutes * 60 +
                t.seconds +
                t.milliseconds / 1000
            )

        for sub in subs:
            start = to_seconds(sub.start)
            end = to_seconds(sub.end)

            if current_start is None:
                current_start = start

            if start - current_start >= chunk_length:
                chunks.append(
                    (current_start, last_end, " ".join(current_text))
                )
                current_start = start
                current_text = []

            current_text.append(sub.text.replace("\n", " "))
            last_end = end

        if current_text:
            chunks.append(
                (current_start, last_end, " ".join(current_text))
            )

        return chunks
```

```python
import json
from collections import defaultdict
import math

INPUT_FILE = "VIDEO_ID.en.json3"
CHUNK_SECONDS = 20

with open(INPUT_FILE, "r", encoding="utf-8") as f:
    data = json.load(f)

chunks = defaultdict(list)

for event in data["events"]:
    if "segs" not in event:
        continue

    start_time = event["tStartMs"] / 1000
    chunk_index = math.floor(start_time / CHUNK_SECONDS)

    text = "".join(seg.get("utf8", "") for seg in event["segs"])
    text = text.replace("\n", " ").strip()

    if text:
        chunks[chunk_index].append(text)

# Print chunks
for idx in sorted(chunks):
    start = idx * CHUNK_SECONDS
    end = start + CHUNK_SECONDS
    combined = " ".join(chunks[idx])

    print(f"[{start:>4}-{end:>4}s] {combined}")
```
